```python
# 2.9.2) через DataFrame.apply
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
column_names = ['Class', 'Alcohol', 'Malic Acid', 'Ash', 'Alcalinity of Ash', 'Magnesium', 'Total Phenols', 'Flavanoids', 
                'Nonflavanoid Phenols', 'Proanthocyanins', 'Color Intensity', 'Hue', 'OD280/OD315 of Diluted Wines', 'Proline']
df1 = pd.read_csv(url, header=None, names=column_names)
df1.columns = column_names
def get_color_intensity_level(row):
    if row['Color Intensity'] < 5:
        return 'Low'
    else:
        return 'High'
    
df1['Color_Intensity_Level'] = df.apply(get_color_intensity_level, axis=1)
df1
```

```python
# 2.9.3) через Series.apply
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
column_names = ['Class', 'Alcohol', 'Malic Acid', 'Ash', 'Alcalinity of Ash', 'Magnesium', 'Total Phenols', 'Flavanoids', 
                'Nonflavanoid Phenols', 'Proanthocyanins', 'Color Intensity', 'Hue', 'OD280/OD315 of Diluted Wines', 'Proline']
df1 = pd.read_csv(url, header=None, names=column_names)
df1['Alcohol'] = df1['Alcohol'].astype(int)
alc_list = list(df1['Alcohol'])
a = pd.Series(alc_list)

def get_alcohol_level(alcohol):
    if alcohol < 12:
        return 'Low'
    elif alcohol >= 12 and alcohol < 14:
        return 'Medium'
    else:
        return 'High'

a.apply(get_alcohol_level)
```
